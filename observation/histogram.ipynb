```python
import matplotlib.pyplot as plt
import numpy as np

data = [[0, 0, scene_proportion13], [scene_proportion4, scene_proportion5, scene_proportion6], [scene_proportion10, scene_proportion11, scene_proportion12], [scene_proportion7, scene_proportion8, scene_proportion9], [scene_proportion1, scene_proportion2, scene_proportion3]]

colors = ['blue', 'orange', 'green'] 
color_labels = ['1/3', '2/3', '1']

bar_width = 1
gap_width = 1
positions = []
current_position = 0

all_values = []
all_colors = []
for i, group in enumerate(data):
    if i == 0:  
        positions.append(current_position) 
        positions.extend([current_position, current_position + 1])  
    else:
        positions.extend(np.arange(current_position, current_position + len(group)))  
    all_values.extend(group)  
    all_colors.extend([colors[j % len(colors)] for j in range(len(group))])  
    current_position += len(group) + gap_width 

plt.figure(figsize=(6, 3))
bars = plt.bar(positions, all_values, width=bar_width, color=all_colors)

handles = [plt.Line2D([0], [0], color=color, lw=4) for color in colors]
plt.legend(handles, color_labels) 

xticks_positions = [1, 5, 9, 13, 17]
xticks_labels = ['no', 'common', 'special', 'fake endpoint', 'mixture']
plt.xticks(xticks_positions, xticks_labels)

plt.ylabel('ASR')

plt.show()
```
